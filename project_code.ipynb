```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

## Model1

```python
def vanderpol_model(arr, time):
    y = arr[0]
    z = arr[1]
    dydt = z
    dzdt = -k*(y*y - 1)*z - y
    return [dydt, dzdt]
```

### x vs dx/dt for different mu values given a fixed init value for x and dx/dt

```python


time = np.linspace(0.0, 50.0, 500)
init = [0.5,1.5]
k_values = [0.025,0.5,4,20]
for i in range(4):
    k = k_values[i]
    ans = odeint(vanderpol_model,init,time)
    plt.plot(ans[:,0], ans[:,1])
    plt.xlabel('x')
    plt.ylabel('dx/dt')
    plt.title('Limit cycle (mu = ' + str(k) + ')')
    plt.show()
    #plt.savefig("./limitcycle_a3.png")
```

### x vs dx/dt for different mu values given a second fixed init value for x and dx/dt

```python
time = np.linspace(0.0, 50.0, 500)
init = [3,3]
k_values = [0.025,0.5,4]
for i in range(3):
    k = k_values[i]
    ans = odeint(vanderpol_model,init,time)
    plt.plot(ans[:,0], ans[:,1])
    plt.xlabel('x')
    plt.ylabel('dx/dt')
    plt.title('Limit cycle (mu = ' + str(k) + ')')
    plt.show()
    #plt.savefig("./limitcycle_a3.png")
```

### x vs t for different mu values at a given initial value for x and dx/dt

```python
time = np.linspace(0.0, 100.0, 500)
init = [0.7,0.7]
k_values = [0,0.1,4, 10, 20]
ans = [0,0,0,0,0]
for i in range(5):
    k = k_values[i]
    ans[i] = odeint(vanderpol_model,init,time)
    plt.plot(time,ans[i][:,0])
    plt.xlabel('t')
    plt.ylabel('x')
    plt.title('x vs t at mu = ' + str(k))
    plt.show() 
```

## Lienard theorem (Model2)

```python
def vanderpol_model2(arr, time):
    y = arr[0]
    z = arr[1]
    dzdt = y/k
    dydt = k*(y - (y*y*y)/3.0 - z)
    return [dydt, dzdt]
```

### x vs y for different mu values at a fixed initial value for x and dx/dt

```python
time = np.linspace(0.0, 300.0, 2000)
init = [0.2,0.2]
k_values = [0.025, 0.5,4,20]
for i in range(4):
    k = k_values[i]
    ans = odeint(vanderpol_model2,init,time)
    plt.plot(ans[:,0], ans[:,1])
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Limit cycle (mu = ' + str(k) + ')')
    plt.show()
    #plt.savefig("limitcycle23_tight.png")
```

### x vs t for different mu values at a fixed initial value for x and dx/dt

```python
time = np.linspace(0.0, 50.0, 500)
init = [1.0,2.0]
k_values = [0.5,1,2,4,8]
for i in range(5):
    k = k_values[i]
    ans = odeint(vanderpol_model2,init,time)
    plt.plot(time,ans[:,0])
    plt.xlabel('t')
    plt.ylabel('x')
    plt.title('x vs t at k = ' + str(k))
    plt.show() 
```

### x vs t for different mu values at a fixed initial value for x and dx/dt in a single figure for comparison using model1
